```python
# This program declares class "Computer". Single line comment

'''
Group Members:
1.Andrew Oropeza
2.Andres Pacho
3.Ashley Palencia
4.Karen Ocampo
Program name: GP-Computer-v10.py
'''

#--------------------  start of class Computer  --------------------
'''
    Class Variable:
    objPartAdded: It keeps track of number of instantiated objects.
                  Shared by all instances.
    objPartDeleted: It keeps track of number of deleted objects.
                    Both class variables shared by all instances of the class, 'Computer'

    Attributes:
    objBrand: Computer brand (e.g., Dell)
    objBoard: Motherboard (e.g., ASUS ROG Strix B450-F Gaming II)
    objCpu: CPU (e.g., Ryzen 7 5800x)
    objRam: Ram (e.g., G.Skill Trident Z Neo Series 32GB)
    Each instance has its own attributes
'''
'''
    Methods:
    __init__(): To initialize attributes
    __del__(): To destroy objects

    Accessor methods:
    getObjBrand(): To access the attribute objBrand
    getObjBoard(): To modify the attribute objBoard
    getObjCpu(): To access the attribute objCpu
    getObjRam(): To access the attribute objRam

    Mutator methods:
    setObjBrand(): To modify the attribute objBrand
    setObjBoard(): To modify the attribute objBoard
    setObjCpu(): To modify the attribute objCpu
    setObjRam(): To modify the attribute objRam

    Other method:
    display(): To display object
    displayCount(): It prints the number of instantiated/deleted
'''

class Computer:

    # class variable to keep track of number of instantiated/deleted objects. Shared by all instances
    objPartAdded = 0
    objPartDeleted = 0
    # constructor
    def __init__(self, objBrand='None', objBoard='None', objCpu='None', objRam='None'):
        self.objBrand = objBrand
        self.objBoard = objBoard
        self.objCpu = objCpu
        self.objRam = objRam
        Computer.objPartAdded += 1

    # destructor
    def __del__(self):
        Computer.objPartDeleted += 1
        print('The object with the brand: ', self.objBrand, 'Motherboard: ', self.objBoard, 'CPU: ' \
              , self.objCpu, 'Ram: ', self.objRam, 'was destroyed.', Computer.objPartDeleted, 'object(s) deleted')

    # accessor methods(s) --------------------------------
    def getObjBrand(self):
        return self.objBrand

    def getObjBoard(self):
        return self.objBoard

    def getObjCpu(self):
        return self.objCpu

    def getObjRam(self):
        return self.objRam

    # mutator methods ------------------------------------
    def setObjBrand(self, objBrand):
        self.objBrand = objBrand
        return self.objBrand

    def setObjBoard(self, objBoard):
        self.objBoard = objBoard
        return self.objBoard

    def setObjCpu(self, objCpu):
        self.objCpu = objCpu
        return self.objCpu

    def setObjRam(self, objRam):
        self.objRam = objRam
        return self.objRam

    # other method(s) ------------------------------------
    def display(self):
        print('Computer brand: ', self.objBrand, '\t','Motherboard: ', \
              self.objBoard,'\t', 'CPU: ', self.objCpu, 'Ram: ', self.objRam, '.', Computer.objPartAdded, 'object(s) instantiated.')

    def displayCount(self):
        print('\nTotal Objects Instantiated so far: ', Computer.objPartAdded,'\n')
    # end of class Computer
#------------------------------------   end of class Computer  ------------------------------------


#------------------------------------   test the class  -------------------------------------------
# This instantiates first set of objects
# default constructor
obj1 = Computer()
obj1.display()

# non-default constructor
obj2 = Computer('Dell', 'ASUS ROG Strix B450-F Gaming II')
obj2.display()

# non-default constructor
obj3 = Computer('Dell', 'ASUS ROG Strix B450-F Gaming II', 'Ryzen 7 5800x')
obj3.display()

# non-default constructor
obj4 = Computer('Dell', 'ASUS ROG Strix B450-F Gaming II', 'Ryzen 7 5800x', 'G.Skill Trident Z Neo Series 32GB')
obj4.display()

# This instantiates second set of objects
# non-default constructor
obj5 = Computer('HP', 'ASUS ROG Strix B450-F Gaming II')
obj5.display()

# non-default constructor
obj6 = Computer('HP', 'ASUS ROG Strix B450-F Gaming II', 'Ryzen 7 5800x')
obj6.display()

# non-default constructor
obj7 = Computer('HP', 'ASUS ROG Strix B450-F Gaming II', 'Ryzen 7 5800x', 'G.Skill Trident Z Neo Series 32GB')
obj7.display()

# displayCount () shows the total number of objects instnatiated so far.
obj7.displayCount()

# using accessor methods
print('----------Note the effect of accessor and mutator methods below----------')

print('\nUsing accessor method for the object obj7 - the value of objBrand = ' , obj7.getObjBrand())
print('Using accessor method for the object obj7 - the value of objBoard = ' , obj7.getObjBoard())
print('Using accessor method for the object obj7 - the value of objCpu = ' , obj7.getObjCpu())
print('Using accessor method for the object obj7 - the value of objRam = ' , obj7.getObjRam(),'\n')

# using mutator methods
print('\nUsing mutator method for the object obj7 - the new value objBrand =  ',
      obj7.setObjBrand('dell'))
print('Using mutator method for the object obj7 - the new value objBoard = ', \
      obj7.setObjBoard('asus rog strix b450-f gaming II'))
print('Using mutator method for the object obj7 - the new value objCpu = ', \
      obj7.setObjCpu('ryzen 7 5800x'))
print('Using mutator method for the object obj7 - the new value objRam = ', \
      obj7.setObjRam('g.skill trident z neo series 32gb'),'\n')


print('----------Note the effect of accessor and mutator methods above----------')

# destroying objects
print('\n----------Starting destroying objects----------\n')
del obj1
del obj2
del obj3
del obj4
del obj5
del obj6
del obj7
```
